## Deploy

1. create a folder where to work temporarily
```
mkdir <TEMPORARY_FOLDER_NAME>
```
2. init github (?: can avoid this step)
```
git init
```
3. optional: create the repository in github
(**CANNOT BE DONE WITH CMD LINE :(**)
```
..
```
4. connect with repository
```
git remote add origin <GITHUB_REPOSITORY_URL>
```
5. pull
```
git pull origin master
```

6. open jupyter notebook
```
jupyter notebook
```

# Save
1. close jupyter
```
ctrl+C
```

2. add all files
```
git add *
```
3. commit new changes
```
git commit -m "<NEW_CHANGES_DONE_MESSAGE>"
```
4. push 
```
git push -u origin master
```
-u:= remeber option origin master, later you can only type 
```
git push
```
5. Get out of the folder
```
cd ..
```
6. Delete Temporary folder
```
rm -rf <TEMPORARY_FOLDER_NAME>
```
