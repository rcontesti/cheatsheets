## 2.1 Definition of the probability space or probability measure $(\Omega, \mathscr{F}, \mathbb{P})$ triple: 
* $\Omega$:= Sample Space: Non empty set.
* $\mathscr{F}$:= Sigma Algebra: Collection of subsets of $\Omega$, containing $\Omega$ and $\varnothing$ and closed under the formation of complements and countable unions and countable interactions $^{1}$
* $\mathbb{P}$:= is a mapping from $\mathscr{F}$ to $[0,1]$ with $\mathbb{P}(\varnothing)=0$ and $\mathbb{P}(\Omega)=1$.


Intuitively $\mathscr{F}$ contains all the subsets of $\Omega$ where $\mathbb(P)$ is well defined but not all the subsets necesarily.

$^{1}$ Means:
1. $\forall$  $A\subseteq \Omega$ if $A\in\mathscr{F}$ then $A^{C}\in\mathscr{F}$
2. $\forall$ countable collection $ \{ A_{i} \}_{i \in I}\subseteq \Omega$ if $A_{i}\in\mathscr{F}$ $\forall _{i}\rightarrow\cup\ A_{i} \in \mathscr{F}$
3. $\forall$ countable collection $ \{ A_{i} \}_{i \in I}\subseteq \Omega$ if $A_{i}\in\mathscr{F}$ $\forall _{i}\rightarrow\cap\ A_{i} \in \mathscr{F}$
